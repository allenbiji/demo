```python
df=pd.read_csv('Bank_data.csv')
df.head()
```

```python
df["y"]=df["y"].map({"yes":1,"no":0})
df.head()
```

```python
x1=df.drop(["Unnamed: 0","y"],axis=1)
x1.head()
y=df["y"]
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
x
```
